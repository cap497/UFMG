```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def infect(k, SIR):
  prob = distribuicao()
  if prob < PC:
    if (k+1) % L == 0:
      ind = k+1-L
    else:
      ind = k+1
    if rede[ind//L][ind%L] == 0 and SIR[0] > 0:
      rede[ind//L][ind%L] = 1
      SIR[0] = SIR[0] - 1
      SIR[1] = SIR[1] + 1

  prob = distribuicao()
  if prob < PC:
    if k > (N-L-1):
      ind = k+L-N
    else:
      ind = k+L
    if rede[ind//L][ind%L] == 0 and SIR[0] > 0:
      rede[ind//L][ind%L] = 1
      SIR[0] = SIR[0] - 1
      SIR[1] = SIR[1] + 1

  prob = distribuicao()
  if prob < PC:
    if k % L == 0:
      ind = k+L-1
    else:
      ind = k-1
    if rede[ind//L][ind%L] == 0 and SIR[0] > 0:
      rede[ind//L][ind%L] = 1
      SIR[0] = SIR[0] - 1
      SIR[1] = SIR[1] + 1

  prob = distribuicao()
  if prob < PC:
    if k < L:
      ind = k+N-L
    else:
      ind = k-L
    if rede[ind//L][ind%L] == 0  and SIR[0] > 0:
      rede[ind//L][ind%L] = 1
      SIR[0] = SIR[0] - 1
      SIR[1] = SIR[1] + 1

  return SIR

def recuperate(k, SIR):
  prob = distribuicao()
  if prob < PR and SIR[1] > 0:
    rede[k//L][k%L] = -1
    SIR[1] = SIR[1] - 1
    SIR[2] = SIR[2] + 1
  return SIR

def status(N, rede):
  return rede[N//L][N%L]

def plotar(SIRs):
  Ss = [s[0] for s in SIRs]
  Is = [i[1] for i in SIRs]
  Rs = [r[2] for r in SIRs]
  Xs = [x for x in range(len(SIRs))]
  f, ax = plt.subplots(1, figsize=(10,8))
  ax.plot(Xs, Ss)
  ax.plot(Xs, Is)
  ax.plot(Xs, Rs)
  f.legend(['Suscet√≠veis','Infectados','Recuperados'], loc='center right')
  plt.title('Pc:' + str(PC) + ' Pr:' + str(PR))

def transpor(M):
  return list(map(list, zip(*M)))
```

```python
Probs = [(0.2, 0.2), (0.2, 0.5), (0.2, 0.8), (0.5, 0.2), (0.5, 0.5), (0.5, 0.8), (0.8, 0.2), (0.8, 0.5), (0.8, 0.8)]
L = 100
N = L ** 2
steps = 10
distribuicao = lambda : np.random.random()
palette = "YlGnBu"

rec = -1 # recuperado
suc = 0  # suscetivel
inf = 1  # infectado
```

```python
all_SIR_series = []

for prob in Probs:
    PC, PR = prob
    rede = np.zeros((L, L))
    rand = np.random.randint(0, N)
    rede[rand // L][rand % L] = 1
    SIR = [N - 1, 1, 0]
    SIR_series = []

    num_heatmaps = (steps + 2) // 3

    plt.figure(figsize=(8 * num_heatmaps, 6))

    for step in range(steps):
        SIR_series.append(list(SIR))

        if step % 3 == 0:
            plt.subplot(1, num_heatmaps, step // 3 + 1)
            sns.heatmap(rede, vmin=-1, vmax=1, cmap=palette)

        for i in range(N):
            SIR = infect(i, SIR)
            if status(i, rede) == inf:
                SIR = recuperate(i, SIR)

    plt.show()
    all_SIR_series.append(SIR_series)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
axes = axes.flatten()

for i, ax in enumerate(axes):
    if i < len(all_SIR_series) and len(all_SIR_series[i]) > 0:
        SIR_series = np.array(all_SIR_series[i]).T
        ax.plot(SIR_series[0], label='Suscet√≠vel')
        ax.plot(SIR_series[1], label='Infectado')
        ax.plot(SIR_series[2], label='Recuperado')
        ax.legend()
        ax.set_title(f"PC = {Probs[i][0]}, PR = {Probs[i][1]}")
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()
```
